```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TRAIN_SIZE = 36000
TEST_SIZE = 4000
```

```python
df = pd.read_csv('output_{}_{}.csv'.format(TRAIN_SIZE, TEST_SIZE))
```

```python
ax = df.plot(grid=True, xlim=(0,100), ylim=(0,1))
ax.set(xlabel='epoch', ylabel='accuracy')
```

```python
fig = ax.get_figure()
fig.savefig('fig_{}_{}.png'.format(TRAIN_SIZE, TEST_SIZE))
```
